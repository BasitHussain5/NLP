```python
from googletrans import Translator

# Create an instance of the Translator
translator = Translator(service_urls=['translate.google.com'])

# Define a function to translate the text to Urdu
def translate_to_urdu(text):
    try:
        # Translate the text to Urdu
        translated_text = translator.translate(text, src='ur', dest='fa').text
        return translated_text
    except Exception as e:
        print(f"Translation error: {e}")
        return text

# Apply the translation function to the "Comment" column
per_data6['comment_persian']  = per_data6['comment'].apply(translate_to_urdu)
```

```python
dataframes = ['per_data1', 'per_data2', 'per_data3','per_data4','per_data5','per_data6']

# Initialize an empty list to store the DataFrames
dfs = []

# Iterate over the DataFrame names and append the DataFrames to the list
for dataframe_name in dataframes:
    df = globals()[dataframe_name]  # Access the DataFrame using the name as a variable
    dfs.append(df)

# Concatenate the DataFrames into a new DataFrame
urdu_to_persian = pd.concat(dfs)

```

```python
from langdetect import detect

# Assuming your dataset is stored in a DataFrame called 'df'

# Function to check if text is in English
def is_english(text):
    try:
        return detect(text) == 'fa'
    except:
        return False

# Filter the DataFrame to include only rows with English text
persian_data = urdu_to_persian[urdu_to_persian['comment_persian'].apply(is_english)]

# Print the resulting filtered DataFrame
```
