```python
import pandas as pd
import numpy as np

# Define the NER function
def ner(text):
    entities = []
    for brand, keywords in ner_dictionary.items():
        for keyword in keywords:
            if keyword in text:
                entities.append((keyword, brand))
                break
    return entities if entities else np.nan

# Load the data into a DataFrame
data = df1.copy()

# Apply the NER function to the "Comment" column to extract brand entities
data['brand'] = data['comment'].apply(ner)
data['brand'] = data['brand'].apply(lambda x: ', '.join([brand[1] for brand in x]) if isinstance(x, list) else x)
data['brand'] = data['brand'].str.split().str[0]
```

```python
import re
import emoji
def remove_emojis(text):
    clean_text = emoji.demojize(text)
    clean_text = re.sub(r'"', '', text)
    clean_text = re.sub(r'[!"#$%&\'()*+,-./:;<=>?@\[\\\]^_{|}~"]', '', text)
    clean_text = re.sub(r'<a href=|"- <a href="', '', text)
    clean_text = re.sub(r'[^a-zA-Z0-9\s,.\'"]', '', text)
    clean_text = re.sub(r'\[.*?\]', '', clean_text)
    clean_text = re.sub(r'[<>\\/[\]]', '', clean_text)
    clean_text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    clean_text = re.sub(r':[^:\s]*(?:::[^:\s]*)*:', '', clean_text)
    clean_text = re.sub(r'[<>]', '', text)
    clean_text = re.sub(r'\s+', ' ', clean_text).strip()
    return clean_text
data1['comment'] = data1['comment'].apply(remove_emojis)
```

```python
def guess_gender_from_username(username):
    username = username.lower()
    male_prefixes = ['mr', 'master', 'mister']
    female_prefixes = ['mrs', 'miss', 'ms', 'lady']
    male_suffixes = ['man', 'guy', 'boy']
    female_suffixes = ['woman', 'girl']
    for prefix in male_prefixes:
        if username.startswith(prefix):
            return 'male'
    for prefix in female_prefixes:
        if username.startswith(prefix):
            return 'female'
    for suffix in male_suffixes:
        if username.endswith(suffix):
            return 'male'
    for suffix in female_suffixes:
        if username.endswith(suffix):
            return 'female'
    pass
usernames = ['john_doe', 'mrs_smith', 'girl_power', 'superman', 'emma_watson']

for username in usernames:
    gender = guess_gender_from_username(username)
    print(f"Username: {username}, Gender: {gender}")
```
